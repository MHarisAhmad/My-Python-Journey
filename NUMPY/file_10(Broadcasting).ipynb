Row broadcasting

```python
ab = np.array([[10,11,12],
              [21,22,23]])
row = np.array([[1,1,1],
               [1,1,1]])
print(ab + row)
```

Column Broadcasting

```python
matrix = np.array([[1,1],
                   [1,1],
                   [1,1]])
matrix.shape
```

```python
column = np.array([10,10,10]).reshape(-1,1)
column.shape
```

```python
print(matrix + column)
```
