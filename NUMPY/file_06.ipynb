```python
import numpy as np
```

```python
x = np.array([1,2,3,4], dtype=np.int32)
print(x.dtype)
```

```python
y = np.array([1,2,3])
z = np.float32(y)
print(z)
print(y.dtype)
print(y)
a = np.int_(z)
print(a.dtype)
```

```python
f = np.array(["a", "b", "c"])
print(f)
print(f.dtype)
```

```python
g = np.array([1,2,3])
print(g)
print(g.dtype)
g = g.astype(float)
print(g)
print(g.dtype)
```
