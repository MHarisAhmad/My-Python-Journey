```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame(
    {
        "month":["Jan","Feb", "Mar", "Apr"],
        "Sales": [100, 120, 90, 140],
        "Profit": [20, 30, 15, 40]
    }
)
```

```python
data.plot(x = "month", y = "Sales")
plt.show
```

```python
data.plot(x="month", y="Sales", kind="bar")
plt.show()
```

```python
data.plot(x="Sales", y="Profit", kind="scatter")
plt.show()
```

```python
data["Sales"].plot(kind="hist")
plt.show()
```

```python
data.plot(kind="box")
plt.show()
```

```python
data.set_index("month")["Sales"].plot(kind="pie", autopct="%1.1f%%")
plt.show()
```

```python
axs = data.plot.area(figsize=(12, 4), subplots=True)

plt.show()
```
